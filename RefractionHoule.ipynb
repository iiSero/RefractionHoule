# A. Modélisation de la réfraction de la houle par tracé de rayons

On s'intéresse à la réfraction de la houle par une bathymétrie. En supposant que les longueurs d'onde des ondes de surface sont petites devant l'échelle de variation de la profondeur, on effectue un tracé de rayons.

## A.1 Bathymétrie

On définit la profondeur de par l'expression
$$h(x,y) =h_{large} -
  \left[1+ \mu \, \sin\left( {2\, \pi\, x \over D}\right)\right]
  \,  \left[ \lambda \, (y-y_{plat}) + \sqrt{\lambda^2 \, (y-y_{plat})^2 + 4} \right]/2\,$$
telle que $h\sim h_{large}$ pour $y\ll y_{plat}$ et où $\lambda$ est la pente moyenne en $y$ du fond, modulée en $x$ avec le facteur $\mu$. La cote du fond est alors $Z_f(x,y) = - h(x,y)$ (iso-profondeurs en pointillés sur la figure 1). La ligne $Z_f(x,y)=0$ (en trait plein) définit le trait de côte.

## A.2 Relation de dispersion des ondes de surface

Lorsque la profondeur $h$ est constante, les ondes de surface planes et monochromatiques de la forme
$$u(x,y,t) = A \, \cos(\underline k \cdot \underline x - \omega \, t + \phi)\;,$$
 de vecteur d'onde   $\underline k = k_{x} \, \underline e_x + k_{y} \, \underline e_y$, sont solutions des équations d'Euler irrotationnelles si et seulement (non détaillé ici) la relation de dispersion suivante est vérifiée :
$$ \omega  = \Omega(k_x, k_y) =\Omega(\underline k)=  \Omega_{houle} (k) = \sqrt{g\, k \tanh(k\, h)} \;,$$
où $g$ est la gravité et $k= \| \underline k\| = \sqrt{k_x^2 + k_y^2}$.
On en déduit les expressions suivantes des vitesses de phase $\underline c_\varphi$ et de groupe $\underline c_g$ :
\begin{eqnarray}
\underline c_\varphi(\underline k)  &= {\Omega(\underline k)\over k} \,\underline e_k=  c_\varphi(k) \, \underline e_k
 & \hbox{avec} \quad
c_\varphi(k) = \sqrt{{g\over k} \, \tanh(k\, h)} \;, \cr
\underline c_g(\underline k) &= {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline k}} \Omega(\underline k) =  c_g(k) \, \underline e_k
 & \hbox{avec} \quad
c_g(k)  ={1\over 2} \,  c_\varphi(k) \, \left[ 1 + {2\, k \, h \over \sinh(2\, k\, h)} \right]  \;,
\end{eqnarray}
où ${\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline k}}  = {\partial \over \partial k_x} \, \underline e_x + {\partial \over \partial k_y} \, \underline e_y $ est le gradient par rapport aux variable $k_x$ et $k_y$.

## Relations de dispersion locale pour une bathymétrie non homogène

Lorsque l'échelle de variation de la profondeur $h(x,y)$ est grande devant la longueur d'onde $L_0 = 2\, \pi / k_0$, avec $k_0 = \| \underline k_0\|$, on considère des paquets d'onde de la forme $u(x,y,t) = A(x,y)\, \cos[\varphi(x,y,t)]$ et on définit les champs suivants :
$$  k_x(x,y,t) = {\partial \varphi\over \partial x} (x,y,t)
  \;, \quad
  k_y(x,y,t) = {\partial \varphi\over \partial y}  (x,y,t)
   \quad \hbox{et} \quad
  \omega(x,y,t) =- {\partial \varphi\over \partial t} (x,y,t)\;.$$
L'approximation WKB (non détaillée ici) permet d'affirmer que la relation de dispersion locale suivante est vérifiée :
 $$ \omega = \Omega(k_x, k_y, x, y)
  \quad \Longleftrightarrow \quad
{\partial \varphi\over \partial t} + \Omega\left( {\partial \varphi\over \partial x} , {\partial \varphi\over \partial y}  , x, y\right) = 0\;,$$
où $\Omega(\underline k, \underline x)$ est obtenu en tenant compte de la variation de $h(\underline x)$ dans la relation de dispersion homogène.  

En prenant le gradient en $\underline x$ de cette équation de l'Eikonale, qui s'écrit  $ {\partial \varphi\over \partial t}  + \Omega(  {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}}\,\varphi, \underline x) =0$, et en utilisant la relation ${\partial k_x\over \partial y}={\partial k_y\over \partial x}={\partial^2 \varphi\over \partial x \, \partial y}$, on obtient une équation de transport pour $\underline k$ qui s'écrit :
$$ {\partial \underline k \over \partial t}  + \underline c_g(\underline k,\underline x) \cdot  {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}}\,   \underline k = - {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}}_{\underline x}\,\Omega(\underline k,  \underline x)\,$$
où $\underline c_g \cdot {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}} = c_{gx} \, {\partial \over \partial x} + c_{gy} \, {\partial \over \partial y}$ est l'opérateur de dérivation selon la vitesse $\underline c_g$ avec $c_{gx} = {\partial \Omega \over \partial k_x}$ et $c_{gy} = {\partial \Omega \over \partial k_y}$, et ${\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline x}} \Omega(\underline k, \underline x)= {\partial \Omega \over \partial x}\, \underline e_x + {\partial \Omega \over \partial y}\, \underline e_y$.

## A.3 Onde plane incidente

On suppose qu'une onde plane arrive d'une zone où la profondeur $h_{large}$ est constante, en faisant un angle $\theta$ avec l'axe $Oy$. L'élévation de la surface de la mer s'écrit alors $u(x,y,t) = A_0 \, \cos(\underline k_0 \cdot \underline x - \omega_0 \, t + \phi_0)$ avec $\underline x = x \, \underline e_x + y \, \underline e_y$ et $\underline k_0 = k_{0x} \, \underline e_x + k_{0y} \, \underline e_y$ où $\underline e_x$ et $\underline e_y$ sont les vecteurs de base, la relation de dispersion $\omega_0 =\sqrt{g\, k_0 \tanh(k\, h_{large})}$ étant vérifiée. On suppose connue cette expression sur un segment de droite $\cal L$ perpendiculaire au vecteur d'onde $\underline k_0$ et situé suffisamment loin de la côte (ligne verte en gras sur la figure 1).

## A.4 Tracé de rayons

La théorie des caractéristique (non détaillée ici), permet de résoudre l'équation de transport du vecteur d'onde $\underline k (\underline x, t)$ en calculant une famille de trajectoires $\underline  x_{\cal C}(t)=x_{\cal C}(t) \, \underline e_x + y_{\cal C}(t) \, \underline e_y$ de l'espace transportant le vecteur $\underline k_{\cal C}(t) = k_{x_{\cal C}}(t) \, \underline e_x + k_{y_{\cal C}}(t) \, \underline e_y$, solutions du  système d'équations différentielles ordinaires (EDO) suivant :
$$ {d \underline x_{\cal C} \over dt }  ={\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline k}}\,\Omega(\underline k_{\cal C}, \underline x_{\cal C})  = \underline c_g(\underline k_{\cal C}, \underline x_{\cal C})
\quad \hbox{et} \quad
 {d \underline k_{\cal C} \over dt }  = - {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline x}}\, \Omega(\underline k_{\cal C}, \underline x_{\cal C})\;.$$
 En utilisant l'expression de la relation de dispersion des ondes de surface, on obtient le sytème suivant d'équations différentielles ordinaires (EDO) :
\begin{equation}
\begin{matrix}
{d x \over dt }  & =
 {1\over 2} \,  {\sqrt{g\, k  \, \tanh(k\, h)}\over k}\,
 \left[1 + {2\, k \, h \over \sinh(2\, k\, h)} \right]\,{k_x \over k}\;,
 & \quad
 {d k_x \over dt }  & =
 -k {\sqrt{ g\, k  \, \tanh(k\, h)} \over \sinh(2\,  k\, h)} \, {\partial h \over \partial x } (x,y)\;,
 \cr
{dy \over dt }  & =
 {1\over 2} \,   {\sqrt{g\, k  \, \tanh(k\, h)}\over k}\,
 \left[1 + {2\, k \, h \over \sinh(2\, k\, h)} \right]\,{k_y \over k}\;,
 & \quad
 {d k_y \over dt } & =
 -k {\sqrt{ g\, k  \, \tanh(k\, h)} \over \sinh(2\,  k\, h)} \, {\partial h \over \partial y } (x,y)\;.
\end{matrix}
\end{equation}
On considère alors les conditions initiales de ces trajectoires pour tout point $\underline x_0 = x_0 \, \underline e_x + y_0 \, \underline e_y$ du segment de droite $\cal L$ en posant
$$  x(0) = x_0 \;, \quad y(0) = y_0 \;, \quad k_x(0) = k_{x0} \;, \quad  k_y(0) = k_{y0} \;. $$
 La forme hamiltionienne de ce système entraine que $\omega(\underline x, t)$ est constante le long des trajectoires, et donc égale à $\omega_0$ pour tout pour $\underline x$ et tout temps $t$. On détermine le vecteur d'onde $\underline k(\underline x)$ en tout point $\underline x$ par lequel passe une trajectoire $\underline x_{\cal C}(t)$ au temps $t$ en écrivant
 $\underline k(\underline x) =\underline k_{\cal C}(t)$. De même, la phase $\varphi_0(\underline x)$ du paquet d'onde au temps $t=0$ est
  $\varphi_0(\underline x) = \varphi_0[\underline x_{\cal C}(t)]$ avec
\begin{eqnarray}
 \varphi_0[\underline x_{\cal C}(t)]  & =&  \underline k_0\cdot \underline x_0 + \phi_0
 + \int_0^t c_{g_{\cal C}}(t') \, k_{\cal C}(t')\, dt' \;,
 \cr
 \hbox{avec} \quad
 k_{\cal C}(t) = \|\underline k_{\cal C}(t)\|
\quad \hbox{et} &&
  c_{g_{\cal C}} = {1\over 2} \,  {\sqrt{g\, k_{\cal C}  \, \tanh[k_{\cal C}\, h(\underline x_{\cal C})]}\over k_{\cal C}}\,
 \left[1 + {2\, k_{\cal C} \, h(\underline x_{\cal C})  \over \sinh[2\, k_{\cal C}\, h(\underline x_{\cal C})]} \right]\;.
\end{eqnarray}
Comme le segment de droite $\cal L$ est perpendiculaire au vecteur d'onde $\underline k_0$, on peut choisir $\phi_0 = - \underline k_0 \cdot \underline x_0$ pour n'importe quel point $\underline x_0$ de $\cal L$, afin que la phase à $t=0$ soit nulle sur ce segment de droite.

## A.5 Conservation de l'énergie

L'énergie par unité de surface d'une houle monochromatique de la forme $u(x,y,t) = A \, \cos(k_x \, x+ k_y\, y - \omega\, t+ \phi)$ est $E={1\over 2} \, \rho\, g\, A^2$, où $\rho$ est la masse volumique de l'eau. L'amplitude $A$ d'un paquet d'onde de la forme $u(\underline x,t) = A(\underline x,t)\, \cos[\varphi(\underline x,t)]$ émis par l'onde plane monochromatique du segment de droite $\cal L$ ne dépend pas du temps et vérifie
$$
{\rm div} \left\{ A^2(\underline x)\,  \underline c_g[\underline k(\underline x)] \right\}
  = {\partial \over \partial x}\left[ A^2(x,y) \,  c_{gx}  \right] +{\partial \over \partial x}\left[  A^2(x,y)\, c_{gy} \right] = 0\;.$$
En intégrant cette relation sur un domaine $\cal D$ compris entre deux trajectoires $\cal C_1$ et $\cal C_2$ et délimité par deux segments de droites $P_1P_2$ et $Q_1Q_2$ (figure 3), on montre que
  $$ |P_1P_2| \, A_P^2 \;\left( \underline c_{g_P} \cdot \underline e_P\right)
=  |Q_1 Q_2| \, A_Q^2 \;\left(  \underline c_{g_Q} \cdot \underline e_Q\right)\;, $$
où $(A^2_P\, \underline c_{gP})$ et $(A^2_Q\,  \underline c_{gQ})$ sont respectivement les valeurs moyennes de $A^2\, \underline c_{g}$ sur les segments $P_1P_2$ et $Q_1Q_2$, et $\underline e_P$ et $\underline e_Q$ les vecteurs unitaires perpendiculaires à ces segments dans la direction des trajectoires.
La hauteur $H=2\,A$ de la houle augmente donc si les trajectoires des rayons se rapprochent ou si la vitesse de groupe diminue. Les vagues déferlent si la hauteur $H$ dépasse la valeur critique $H_c$ donnée par le critère de déferlement de Miche :
$$H_c = 0,14\, {2\, \pi \over k} \, \tanh(k\, h) \;. $$
Dans ce cas, on annule numériquement la hauteur $H$, ce qui permet de visualiser la ligne de déferlement lors du tracé des iso-contours du champ $H(x,t)$, comme on peut le voir sur la figure 1.

```python
# Sous-progammes de l'énoncé à compiler : ne pas modifier ici

# Import de librairies
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


def inifig(fx,fy,xaxe=0,yaxe=0,tit='Titre',xlab='x',ylab='y'):
    # initialisation d'une figure
    fig=plt.figure(figsize=(fx,fy))
    plt.axvline(x=xaxe)
    plt.axhline(y=yaxe)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.xlabel(xlab,fontsize=16)
    plt.ylabel(ylab,fontsize=16)
    plt.title(tit,fontsize=16)
    plt.minorticks_on()
    plt.grid(which='minor',color='grey', axis='y', linestyle=':', linewidth=1)
    plt.grid(color='black', axis='y', linestyle='-',linewidth=1)
    plt.grid(color='black', axis='x', linestyle='-',linewidth=1)
    return fig


def bathy(x,y):
    # bathymetrie h(x,y)
    kD=2*np.pi/D
    yt=la*(y-yplat); sq=np.sqrt(yt**2+4)
    h=hlarge-(1+mu*np.sin(kD*x))*.5*(yt+sq)
    hx=-mu*kD*np.cos(kD*x)*.5*(yt+sq)
    hy=-(1+mu*np.sin(kD*x))*.5*la*(1+yt/sq)
    return h,hx,hy

def visubathy(x,y):
  # Trace de Zf=-h(x,y)
  X, Y = np.meshgrid(x, y)
  h,Hx,Hy=bathy(X,Y)
  Zf=-h
  # Isolignes
  fig=inifig(fx,fy,tit=r'Tracé de rayons')
  contours = plt.contour(X, Y, Zf, levels=10, colors='black')  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.0f")   # label the lines
  return fig

def visubathy3D(x,y):
  # Trace de  Zf=-h(x,y) en 3D
  X, Y = np.meshgrid(x, y)
  h,Hx,Hy=bathy(X,Y)
  Zf=-h
  # Color map
  # plot 3D
  fig = plt.figure(figsize=(9, 9))
  ax = fig.add_subplot(111, projection='3d')
  # clip Zf
  Zf=np.where(Zf<0, Zf,0)
  Nyh=0;
  Xtron=np.zeros((Nyh,Nx));
  Ytron=np.zeros((Nyh,Nx));
  Zftron=np.zeros((Nyh,Nx));
  Xtron=X[Nyh:Ny,:]; Ytron=Y[Nyh:Ny,:]; Zftron=Zf[Nyh:Ny,:]
  ax.plot_wireframe(Xtron, Ytron, Zftron, color='k', linewidth=0.5)
  ax.view_init(elev=20, azim=-40)  # tilt elev, rotate azim around z-axis
  ax.set_xlim(xmin, xmax)
  ax.set_ylim(ymin, ymax)
  #ax.set_zlim(-15, 0)
  #fig.colorbar(surf, ax=ax, shrink=.6, label=r'$Z_f(x, y)$')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel(r'$Z_f$')
  #ax.set_title(r'Batymétrie $Z_f(x, y)$')
  fig.savefig('visubathy3D.pdf')
  return fig

def bathyclip(x,y,la,mu,D):
    # bathymetrie h(x,y) avec clip en hmin
    h,hx,hy = bathy(x,y)
    h=np.where(h >0, h, hmin)
    hx=np.where(h >0, hx, 0)
    hy=np.where(h >0, hy, 0)
    return h,hx,hy

def houle(k,h):
  # vitesses de phase et de groupe
  kh=k*h
  om=np.sqrt(g*k*np.tanh(kh))
  cphi=om/k;
  cg=.5*cphi*(1+2*kh/np.sinh(2*kh))
  return om, cphi, cg

def visuhoule(h,kmax):
  # visualisation de omega, cphi et cg
  k=np.linspace(.001,kmax,500)
  om, cphi,cg=houle(k,h)
  xla=r'$k$';  yla=r'$\omega$, $c_\varphi$, $c_g$';
  ti=r'Relation de dispersion $\omega= \sqrt{g\, k\, \tanh(k\,h)}$'
  fig=inifig(8.6,5,xlab=xla,ylab=yla,tit=ti)
  plt.plot(k,om,color="green",linewidth=4)
  plt.plot(k,cphi,color="blue",linewidth=4)
  plt.plot(k,cg,color="red",linewidth=4)
  fig.savefig('visuhoule.pdf')

def rayon(x,y,kx,ky,dt,Nt):
  # trace d'un rayon unique
  # initialisation des tableaux 1D
  xr=np.zeros(Nt); yr=np.zeros(Nt);
  kxr=np.zeros(Nt); kyr=np.zeros(Nt);
  cgxr=np.zeros(Nt); cgyr=np.zeros(Nt);
  phir=np.zeros(Nt); hr=np.zeros(Nt);
  phi=kx*x+ky*y+phi0;
  # boucle sur le temps
  for i in range(Nt):
    k=np.sqrt(kx**2+ky**2)
    h,hx,hy=bathyclip(x,y,la,mu,D)
    om,cphi,cg=houle(k,h)
    cgx=cg*kx/k; cgy=cg*ky/k;
    # stockage du rayon dans les tableaux
    xr[i]=x; yr[i]=y;
    kxr[i]=kx; kyr[i]=ky;
    cgxr[i]=cgx; cgyr[i]=cgy;
    phir[i]=phi; hr[i]=h;
    # avance seulement pour h>hmin
    if h>hmin:
      # avance en x,y
      x=x+cgx*dt; y=y+cgy*dt;
      # avance en phi
      phi=phi+cg*k*dt
      # avance en kx, ky
      komsinh=k*om/np.sinh(2*k*h)
      Fkx=-komsinh*hx; Fky=-komsinh*hy;
      kx=kx+Fkx*dt; ky=ky+Fky*dt

  return xr,yr,kxr,kyr,cgxr,cgyr,phir,hr

def rayons(th):
  # calcul de tous les rayons

  # initialisation des tableaux 2D
  xl=np.zeros((Ns+1,Nt)); yl=np.zeros((Ns+1,Nt));
  kxl=np.zeros((Ns+1,Nt)); kyl=np.zeros((Ns+1,Nt));
  cgxl=np.zeros((Ns+1,Nt)); cgyl=np.zeros((Ns+1,Nt));
  phil=np.zeros((Ns+1,Nt)); hl=np.zeros((Ns+1,Nt));

  # boucle sur les points de depart de L
  dsx=ds*np.cos(th); dsy=-ds*np.sin(th)
  for n in range(Ns+1):
    xs=xd+n*dsx; ys=yd+n*dsy;
    # calcul des rayons
    xr,yr,kxr,kyr,cgxr,cgyr,phir,hr=rayon(xs,ys,kx0,ky0,dt,Nt)
    xl[n,:]=xr; yl[n,:]=yr;
    kxl[n,:]=kxr; kyl[n,:]=kyr;
    cgxl[n,:]=cgxr; cgyl[n,:]=cgyr;
    phil[n,:]=phir; hl[n,:]=hr;

  return xl,yl,kxl,kyl,cgxl,cgyl,phil,hl

def visurayons():
  # Visualisation des rayons

  # plots de la bathymetrie et de la ligne d'emission L
  fig=visubathy(x,y);
  plt.plot([xd,xd+2*D],[yd,yd-2*D*np.sin(th)],color="green",linewidth=3)

  # Calcul de tous les rayons
  xl,yl,kxl,kyl,cgxl,cgyl,phil,hl=rayons(th)

  # Energie
  #########
  # initialisation des champs intermediaires
  xm=np.zeros((Ns,Nt));ym=np.zeros((Ns,Nt))
  kxm=np.zeros((Ns,Nt));kym=np.zeros((Ns,Nt))
  km=np.zeros((Ns,Nt));
  cgxm=np.zeros((Ns,Nt));cgym=np.zeros((Ns,Nt))
  phim=np.zeros((Ns,Nt)); hm=np.zeros((Ns,Nt))

  kl=np.zeros((Ns+1,Nt)); # kl est défini ici

  edlxm=np.zeros((Ns,Nt));edlym=np.zeros((Ns,Nt))
  Em=np.zeros((Ns,Nt));

  kl=np.sqrt(kxl**2+kyl**2); # kl est calculé ici

  # calculs aux points intermediaires
  xm=.5*(xl[0:Ns,:]+xl[1:Ns+1,:]);
  ym=.5*(yl[0:Ns,:]+yl[1:Ns+1,:]);
  kxm=.5*(kxl[0:Ns,:]+kxl[1:Ns+1,:]);
  kym=.5*(kyl[0:Ns,:]+kyl[1:Ns+1,:]);
  km=.5*(kl[0:Ns,:]+kl[1:Ns+1,:]);
  cgxm=.5*(cgxl[0:Ns,:]+cgxl[1:Ns+1,:]);
  cgym=.5*(cgyl[0:Ns,:]+cgyl[1:Ns+1,:]);
  phim=.5*(phil[0:Ns,:]+phil[1:Ns+1,:]);
  hm=.5*(hl[0:Ns,:]+hl[1:Ns+1,:]);


  dxm=xl[1:Ns+1,:]-xl[0:Ns,:];
  dym=yl[1:Ns+1,:]-yl[0:Ns,:];
  edlxm=-dym; edlym=dxm;

  # Initialisation de la hauteur de houle
  Em[:,0]=np.zeros(Ns)+E0
  for i in range(1,Nt):
    cgdelim1=cgxm[:,i-1]*edlxm[:,i-1]+cgym[:,i-1]*edlym[:,i-1]
    cgdeli=cgxm[:,i]*edlxm[:,i]+cgym[:,i]*edlym[:,i]
    Em[:,i]=Em[:,i-1]*cgdelim1/cgdeli

  # visualisation hauteur de la houle Hm
  Em=np.where(Em>0, Em, 0)
  Hm=2*np.sqrt(Em);
  Hmc=.14*(2*pi)/km*np.tanh(km*hm) # critere de Miche
  Hm=np.where(Hm < Hmc, Hm, 0)
  contours = plt.contour(xm,ym,Hm,levels=30,colors='brown')  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Phase
  #########
  contours = plt.contour(xl,yl,phil/(2*pi),levels=20,colors='green',linewidths=2)  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Rayons
  #########
  Nsr=Ns//Nr
  for n in range(Ns+1)[::Nsr]:
    # trace des rayons
    xr=xl[n,:]; yr=yl[n,:];
    plt.plot(xr,yr,color="blue",linewidth=1)
    # trace des vitesses de groupe
    Ntq=Nt//10;
    xq=xl[n,::Ntq]; yq=yl[n,::Ntq]; kxq=kxl[n,::Ntq]; kyq=kyl[n,::Ntq];
    cgxq=cgxl[n,::Ntq]; cgyq=cgyl[n,::Ntq]
    plt.quiver(xq,yq,cgxq,cgyq,color='red',scale=100)
    plt.axis('equal')

  ##############
  plt.xlim(xmin,xmax)
  plt.ylim(ymin,ymax)
  fig.savefig('visurayons.pdf')
  fig.savefig('visurayons.jpg')

  return xl,yl,kxl,kyl,kl,cgxl,cgyl,phil,hl, \
         xm,ym,kxm,kym,km,cgxm,cgym,phim,hm,Hm
```

```python
# Sous-programmes pour reconstituer les paramètres par défaut de l'énoncé
# Ces paramètres par défaut ne doivent pas être modifiés ici
# Leur modification doit être effectuée uniquement dans la partie du projet

def pardefaut(Print=True):
  # Paramètres par défaut à ne pas modifier ici
  g=9.81;  # gravité
  pi=np.pi # nombre pi
  D=1000 # période de h en x (défaut 1000)
  hlarge=15; # profondeur au large (défaut 15)
  yplat=-1.5*D; # profondeur environ hlarge si y<<yflat (défaut -1.5*D)
  la=.01 # pente de h en y (défaut .01)
  mu=.1 # amplitude pour la modulation de h (défaut .1)
  th=pi/12 # angle segment de droite L (défaut pi/12)
  L0=100 # longueur d'onde au large (défaut 100)
  H0=1; # hauteur de la houle au large (défaut 1)
  xd=0; # origine x du segment de droite L (défaut 0)
  yd=-2*D # origine y du segment de droite L (défaut -2D)
  Ns=200 # nombre de points de depart sur L (défaut 100)
  Nr=10 # nombre de rayons a tracer (défaut 10)
  dt=2; # pas de temps du schéma temporel (défaut 2)
  Nt=200; # nombre d'itération du schéma temporel (défaut 200)
  hmin=.1 # seuil arrêt pour h (défaut .1)
  Nx=100; # nombre de points en x pour le tracé de h (défaut 100)
  Ny=100;  # nombre de points en y pour le tracé de h (défaut 100)
  xmin=0; # inf en x de de la visualisation (défaut 0)
  xmax=2*D; # sup en x du domaine de la visualisation (défaut 2D)
  ymin=-2*D-2*D*np.sin(th); # inf en y de la visualisation (défaut-2D-2Dsin th)
  ymax=2*mu/la;  # inf en y de la visualisation (défaut 20mu/la)
  fx=8; # taille en x des figures (défaut 8)
  fy=10; # taille en y des figures (défaut 10)

  if Print:
    print("======================")
    print("Paramètres par défaut :")
    print("g=",g)
    print("pi",pi)
    print("D=",D)
    print("hlarge=",hlarge)
    print("yplat=",yplat)
    print("la=",la)
    print("mu=",mu)
    print("th=",th)
    print("L0=",L0)
    print("H0=",H0)
    print("xd=",xd)
    print("yd=",xd)
    print("Ns=",Ns)
    print("Nr=",Nr)
    print("dt=",dt)
    print("Nt=",Nt)
    print("hmin=",hmin)
    print("Nx=",Nx)
    print("Ny=",Ny)
    print("xmin=",xmin)
    print("xmax=",xmax)
    print("ymin=",ymin)
    print("ymax=",ymax)
    print("fx=",fx)
    print("fy=",fy)
    print("======================")

  return g,pi,D,hlarge,yplat,la,mu,th,L0,H0,xd,yd,Ns,Nr, \
         dt,Nt,hmin,Nx,Ny,xmin,xmax,ymin,ymax,fx,fy

def parderives():
  # paramètres dérivés devant être calculés après modification dans le projet
  ds=2*D/Ns
  k0=2*pi/L0;
  kx0=k0*np.sin(th);
  ky0=k0*np.cos(th)
  phi0=-(kx0*xd+ky0*yd);
  E0=(H0/2)**2;
  x=np.linspace(0,xmax,Nx);
  y=np.linspace(ymin,ymax,Ny);
  return ds,k0,kx0,ky0,phi0,E0,x,y
```

## Expérience 01 : Changement de la bathymétrie et de l'incidence

Dans cette expérience, on redéfinit le sous-progamme qui calcule la forme de la bathymétrie avec la commande `bahty=bathy01` dans le programme principal. On modifie aussi le paramètre $\mu$ de sa modulation en $x$ ainsi que l'angle d'incidence de la houle. On trace deux fois plus de rayons.

### 01.a Nouveaux sous-programmes

```python
# Expérience 01 : nouveaux sous-programmes

def bathy01(x,y):
    # bathymetrie h(x,y) de l'énoncé décalée de D/2
    kD=2*np.pi/D
    yt=la*(y-yplat); sq=np.sqrt(yt**2+4)
    h=hlarge-(1+mu*np.sin(kD*(x-D/2)))*.5*(yt+sq)
    hx=-mu*kD*np.cos(kD*(x-D/2))*.5*(yt+sq)
    hy=-(1+mu*np.sin(kD*(x-D/2)))*.5*la*(1+yt/sq)
    return h,hx,hy

```

### 01.b Nouveau programme principal

```python
# Expérience 01 : nouveau programme principal

if Exp01:

  ## BLOC PARAMETRES ###################### Par défaut :
  g,pi,D,hlarge,yplat,la,mu,th,L0,H0,xd,yd,Ns,Nr, \
  dt,Nt,hmin,Nx,Ny,xmin,xmax,ymin,ymax,fx,fy=pardefaut(False)

  # Modification des paramètres ici
  mu=.3; # modulation en x plus grand
  th=pi/24; # angle d'incidence de la houle
  ymax=21*mu/la; # pour voir tout le trait de côte
  Ns=400 # nombre de rayons à calculer
  Nr=20; # nombre de rayons à tracer
  Nt=300; # nombre de pas de temps pour les tracés
  print("mu=",mu,"th=",th,"ymax=",ymax,"Ns=",Ns,"Nr=",Nr,"Nt=",Nt)

  # Paramètres dérivés
  ds,k0,kx0,ky0,phi0,E0,x,y=parderives()
  ############################### FIN BLOC PARAMETRES ##

  # Modification des sous-programmes de l'énoncé
  bathy=bathy01

  # Visualisation de la nouvelle bathymétrie
  visubathy3D(x,y); plt.show();

  # Calcul des rayons
  visurayons()

```

### 01.c Interprétation de l'expérience


* La nouvelle bathymétrie est décalée d'une moitié de longeur par rapport à celle  de l'énoncé.

* La modulation en $x$ étant plus marquée, les baies et les caps sont plus prononcés. Au voisinage des caps, le déferlement est encore plus éloigné de la côte.

* Il a fallu augmenter le nombre d'itérations en temps pour que les rayons atteignent tous le rivage et il a fallu calculer deux fois plus de rayons pour une meilleure précision dans le tracé des iso-phases et des iso-hauteurs de houle dans les baies.

## Expérience 02 : Tracé de la longueur d'onde le long de rayons


Dans cette expérience, on trace la longeur d'onde $2\, \pi/k$ le long de plusieurs rayons.

### 02.a Nouveaux sous-programmes

```python
# Expérience 02 : nouveaux sous-programmes

def visuLong_L02():
    # tracé de la longeur d'onde L le long de rayons

    # Extraction de Nrv rayons parmi Ns
    Nsv=Ns//Nrv;
    # intialisation des tableaux 2D
    scu=np.zeros((Nrv+1,Nt));
    Lr=np.zeros((Nrv+1,Nt));

    # Extraction de Nsv rayons
    for n in range(Ns+Nsv)[::Nsv]:
      nr=n//Nsv  #print("n=",n,"nr=",nr)

      # coordonnée curviligne du rayon
      xrn=xl[n,:]; yrn=yl[n,:];
      for i in range(Nt-1):
        dx=xrn[i+1]-xrn[i]; dy=yrn[i+1]-yrn[i];
        scu[nr,i+1]=scu[nr,i]+np.sqrt(dx**2+dy**2);

      # cacul de k et L
      kxrn=kxl[n,:]; kyrn=kyl[n,:];
      krn=np.sqrt(kxrn**2+kyrn**2);
      Lr[nr,:]=2*pi/krn;

    # Initialisation de la figure pour L
    titre=r"Longueur d'onde $L$ le long des rayons"
    fig=inifig(8,6,tit=titre,xlab=r"coordonnée curviligne $s$",ylab=r"$L$")
    for n in range(Nrv+1):
      line,=plt.plot(scu[n,:],Lr[n,:],linewidth=1)
      Nnum=2*Nt//3; col=line.get_color() # indice où écrire le numéro
      plt.text(scu[n,Nnum],Lr[n,Nnum],str(n),color=col,fontsize=12,va='center')
    fig.savefig('visuLong_L01.pdf')
```

### 02.b Nouveau programme principal

```python
# Expérience 02 : nouveau programme principal

if Exp02:

  ## BLOC PARAMETRES ###################### Par défaut :
  g,pi,D,hlarge,yplat,la,mu,th,L0,H0,xd,yd,Ns,Nr, \
  dt,Nt,hmin,Nx,Ny,xmin,xmax,ymin,ymax,fx,fy=pardefaut()

  # Modification des paramètres ici
  hlarge=30;
  la=.02;
  mu=.08;
  th=pi/24;
  ymax=30*mu/la; # pour voir tout le trait de côte
  print("hlarge=",hlarge,"la=",la,"mu=",mu,"th=",th,"ymax=",ymax)

  # Paramètres dérivés
  ds,k0,kx0,ky0,phi0,E0,x,y=parderives()
  ############################### FIN BLOC PARAMETRES ##

  # Nouveaux paramètres
  Nrv=10 # nombre de rayons à visualiser = Nrv+1

  # Calcul des rayons
  xl,yl,kxl,kyl,kl,cgxl,cgyl,phil,hl, \
  xm,ym,kxm,kym,km,cgxm,cgym,phim,hm,Hm=visurayons()

  # Visualisation de la nouvelle bathymétrie
  visubathy3D(x,y); plt.show();

  # Tracé de la longueur d'onde L le long de rayons
  visuLong_L02()
```

### 02.c Inteprétation de l'expérience

* La longueur d'onde $L=2\,\pi/k$ diminue car la profondeur $h$ diminue. En effet, comme $\omega_0 = \sqrt{g\, k\, \tanh(k\, h)}$ avec $\omega_0$ constant, on peut écrire $h={1\over k}\, {\rm arctg}\, \left({\omega_0^2 \over g\, k}\right)$, qui est une fonction croissante de $k$. Par conséquent, si $h$ diminue $k$ augmente et donc $L$ diminue.

## Expérience 03 : Tracé de la hauteur de houle le long de rayons

Dans cette expérience, on trace la hauteur de houle $H$ le long de plusieurs rayons.


### 03.a Nouveaux sous-programmes

```python
# Expérience 03 : nouveaux sous-programmes

def visuLongH03():
    # tracé de la hauteur de houle le long des rayons

    # Extraction de Nrv rayons
    Nsv=Ns//Nrv; # print("Ns, Nrv, Nsv, Nt", Ns, Nrv, Nsv, Nt)
    Nsvh=Nsv//2

    # intialisation des tableaux 2D
    scu=np.zeros((Nrv,Nt));
    Hr=np.zeros((Nrv,Nt));
    # Extraction de Nsv rayons
    for n in range(Nsvh,Ns+Nsvh)[::Nsv]:
      nr=n//Nsv #print("n=",n,"nr=",nr)
      # coordonnée curviligne du rayon
      xrn=xm[n,:]; yrn=ym[n,:];
      for i in range(Nt-1):
        dx=xrn[i+1]-xrn[i]; dy=yrn[i+1]-yrn[i];
        scu[nr,i+1]=scu[nr,i]+np.sqrt(dx**2+dy**2);
      # cacul de H
      Hr[nr,:]=Hm[n,:];

    # Initialisation de la figure pour H
    titre=r"Hauteur de la houle $H$ le long des rayons"
    fig=inifig(8,6,tit=titre,xlab=r"coordonnée curviligne $s$",ylab="$H$")
    plt.ylim(Hmin,Hmax)
    for n in range(Nrv):
      line,=plt.plot(scu[n,:],Hr[n,:],linewidth=1)
      col=line.get_color() # indice où écrire le numéro
      plt.text(scu[n,Nnum],Hr[n,Nnum],str(n),color=col,fontsize=12,va='center')
    fig.savefig('visuLong_H.pdf')
    plt.show()
```

### 03.b Nouveau programme principal

```python
# Expérience 03 : nouveau programme principal

if Exp03:

  ## BLOC PARAMETRES ###################### Par défaut :
  g,pi,D,hlarge,yplat,la,mu,th,L0,H0,xd,yd,Ns,Nr, \
  dt,Nt,hmin,Nx,Ny,xmin,xmax,ymin,ymax,fx,fy=pardefaut(False)

  # Modification des paramètres ici
  th=pi/48
  ymax=25*mu/la
  print("th=",th,"ymax=",ymax)

  # Paramètres dérivés
  ds,k0,kx0,ky0,phi0,E0,x,y=parderives()
  ############################### FIN BLOC PARAMETRES ##

  # Nouveaux paramètres
  Nrv=10 # nombre de rayons à visualiser
  Hmin=.8*H0; Hmax=2*H0 # limites pour le tracé de H
  Nnum=Nt//2; # indice du tracé des numéros de rayons

  # Calcul des rayons
  xl,yl,kxl,kyl,kl,cgxl,cgyl,phil,hl, \
  xm,ym,kxm,kym,km,cgxm,cgym,phim,hm,Hm=visurayons()

  # Tracé de la hauteur de houle le long des rayons
  visuLongH03()
```

### 03.c Interprétation de l'expérience

Le tracé de la hauteur de la houle indique bien l'augmentation de la hauteur de la houle au voisinage de la côte, jusqu'au déferlement.

### 04.a Nouveaux sous-programmes

```python
# Expérience 04 : nouveaux sous-programme

def visubathy3D_04(x,y):
  # Trace de  Zf=-h(x,y) en 3D
  # Sans modification de la fonction sauf le nom du fichier pour visualiser toutes les bathys des expériences suivantes et/ou précedantes
  X, Y = np.meshgrid(x, y)
  h,Hx,Hy=bathy(X,Y)
  Zf=-h
  # Color map
  # plot 3D
  fig = plt.figure(figsize=(9, 9))
  ax = fig.add_subplot(111, projection='3d')
  # clip Zf
  Zf=np.where(Zf<0, Zf,0)
  Nyh=0;
  Xtron=np.zeros((Nyh,Nx));
  Ytron=np.zeros((Nyh,Nx));
  Zftron=np.zeros((Nyh,Nx));
  Xtron=X[Nyh:Ny,:]; Ytron=Y[Nyh:Ny,:]; Zftron=Zf[Nyh:Ny,:]
  ax.plot_wireframe(Xtron, Ytron, Zftron, color='k', linewidth=0.5)
  ax.view_init(elev=20, azim=-40)  # tilt elev, rotate azim around z-axis
  ax.set_xlim(xmin, xmax)
  ax.set_ylim(ymin, ymax)
  #ax.set_zlim(-15, 0)
  #fig.colorbar(surf, ax=ax, shrink=.6, label=r'$Z_f(x, y)$')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel(r'$Z_f$')
  #ax.set_title(r'Batymétrie $Z_f(x, y)$')
  fig.savefig('visubathy3D_04.pdf')
  return fig

def visurayons04():
  # Visualisation des rayons sans modification de la fonction de base sauf le nom du fichier pour avoir plusieurs fichiers pour chaque expérience (pas forcément la même bathy)

  # plots de la bathymetrie et de la ligne d'emission L
  fig=visubathy(x,y);
  plt.plot([xd,xd+2*D],[yd,yd-2*D*np.sin(th)],color="green",linewidth=3)

  # Calcul de tous les rayons
  xl,yl,kxl,kyl,cgxl,cgyl,phil,hl=rayons(th)

  # Energie
  #########
  # initialisation des champs intermediaires
  xm=np.zeros((Ns,Nt));ym=np.zeros((Ns,Nt))
  kxm=np.zeros((Ns,Nt));kym=np.zeros((Ns,Nt))
  km=np.zeros((Ns,Nt));
  cgxm=np.zeros((Ns,Nt));cgym=np.zeros((Ns,Nt))
  phim=np.zeros((Ns,Nt)); hm=np.zeros((Ns,Nt))

  kl=np.zeros((Ns+1,Nt)); # kl est défini ici

  edlxm=np.zeros((Ns,Nt));edlym=np.zeros((Ns,Nt))
  Em=np.zeros((Ns,Nt));

  kl=np.sqrt(kxl**2+kyl**2); # kl est calculé ici

  # calculs aux points intermediaires
  xm=.5*(xl[0:Ns,:]+xl[1:Ns+1,:]);
  ym=.5*(yl[0:Ns,:]+yl[1:Ns+1,:]);
  kxm=.5*(kxl[0:Ns,:]+kxl[1:Ns+1,:]);
  kym=.5*(kyl[0:Ns,:]+kyl[1:Ns+1,:]);
  km=.5*(kl[0:Ns,:]+kl[1:Ns+1,:]);
  cgxm=.5*(cgxl[0:Ns,:]+cgxl[1:Ns+1,:]);
  cgym=.5*(cgyl[0:Ns,:]+cgyl[1:Ns+1,:]);
  phim=.5*(phil[0:Ns,:]+phil[1:Ns+1,:]);
  hm=.5*(hl[0:Ns,:]+hl[1:Ns+1,:]);


  dxm=xl[1:Ns+1,:]-xl[0:Ns,:];
  dym=yl[1:Ns+1,:]-yl[0:Ns,:];
  edlxm=-dym; edlym=dxm;

  # Initialisation de la hauteur de houle
  Em[:,0]=np.zeros(Ns)+E0
  for i in range(1,Nt):
    cgdelim1=cgxm[:,i-1]*edlxm[:,i-1]+cgym[:,i-1]*edlym[:,i-1]
    cgdeli=cgxm[:,i]*edlxm[:,i]+cgym[:,i]*edlym[:,i]
    Em[:,i]=Em[:,i-1]*cgdelim1/cgdeli

  # visualisation hauteur de la houle Hm
  Em=np.where(Em>0, Em, 0)
  Hm=2*np.sqrt(Em);
  Hmc=.14*(2*pi)/km*np.tanh(km*hm) # critere de Miche
  Hm=np.where(Hm < Hmc, Hm, 0)
  contours = plt.contour(xm,ym,Hm,levels=30,colors='brown')  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Phase
  #########
  contours = plt.contour(xl,yl,phil/(2*pi),levels=20,colors='green',linewidths=2)  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Rayons
  #########
  Nsr=Ns//Nr
  for n in range(Ns+1)[::Nsr]:
    # trace des rayons
    xr=xl[n,:]; yr=yl[n,:];
    plt.plot(xr,yr,color="blue",linewidth=1)
    # trace des vitesses de groupe
    Ntq=Nt//10;
    xq=xl[n,::Ntq]; yq=yl[n,::Ntq]; kxq=kxl[n,::Ntq]; kyq=kyl[n,::Ntq];
    cgxq=cgxl[n,::Ntq]; cgyq=cgyl[n,::Ntq]
    plt.quiver(xq,yq,cgxq,cgyq,color='red',scale=100)
    plt.axis('equal')

  ##############
  plt.xlim(xmin,xmax)
  plt.ylim(ymin,ymax)
  fig.savefig('visurayons04.pdf')
  fig.savefig('visurayons04.jpg')

  return xl,yl,kxl,kyl,kl,cgxl,cgyl,phil,hl, \
         xm,ym,kxm,kym,km,cgxm,cgym,phim,hm,Hm

def bathy04(x,y):
    # bathymetrie h(x,y) avec une gaussienne


    x_centre = 1000
    y_centre = -100

    bosse = 50.0
    h = hlarge - bosse * np.exp(-((x - x_centre)**2 + (y - y_centre)**2) / 300000)
    hx = (bosse/300000)* 2 * (x-x_centre) * np.exp(-((x - x_centre)**2 + (y - y_centre)**2) / 300000)
    hy = (bosse/300000)* 2 * (y-y_centre) * np.exp(-((x - x_centre)**2 + (y - y_centre)**2) / 300000)
    return h,hx,hy


```

```python
def visubathy3D_05(x,y):
  # Trace de  Zf=-h(x,y) en 3D
  # Sans modification de la fonction sauf le nom du fichier pour visualiser toutes les bathys des expériences suivantes et/ou précedantes
  X, Y = np.meshgrid(x, y)
  h,Hx,Hy=bathy(X,Y)
  Zf=-h
  # Color map
  # plot 3D
  fig = plt.figure(figsize=(9, 9))
  ax = fig.add_subplot(111, projection='3d')
  # clip Zf
  Zf=np.where(Zf<0, Zf,0)
  Nyh=0;
  Xtron=np.zeros((Nyh,Nx));
  Ytron=np.zeros((Nyh,Nx));
  Zftron=np.zeros((Nyh,Nx));
  Xtron=X[Nyh:Ny,:]; Ytron=Y[Nyh:Ny,:]; Zftron=Zf[Nyh:Ny,:]
  ax.plot_wireframe(Xtron, Ytron, Zftron, color='k', linewidth=0.5)
  ax.view_init(elev=20, azim=-40)  # tilt elev, rotate azim around z-axis
  ax.set_xlim(xmin, xmax)
  ax.set_ylim(ymin, ymax)
  #ax.set_zlim(-15, 0)
  #fig.colorbar(surf, ax=ax, shrink=.6, label=r'$Z_f(x, y)$')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel(r'$Z_f$')
  #ax.set_title(r'Batymétrie $Z_f(x, y)$')
  fig.savefig('visubathy3D_05.pdf')
  return fig

def visurayons05():
  # Visualisation des rayons sans modification de la fonction de base sauf le nom du fichier pour avoir plusieurs fichiers pour chaque expérience (pas forcément la même bathy)

  # plots de la bathymetrie et de la ligne d'emission L
  fig=visubathy(x,y);
  plt.plot([xd,xd+2*D],[yd,yd-2*D*np.sin(th)],color="green",linewidth=3)

  # Calcul de tous les rayons
  xl,yl,kxl,kyl,cgxl,cgyl,phil,hl=rayons(th)

  # Energie
  #########
  # initialisation des champs intermediaires
  xm=np.zeros((Ns,Nt));ym=np.zeros((Ns,Nt))
  kxm=np.zeros((Ns,Nt));kym=np.zeros((Ns,Nt))
  km=np.zeros((Ns,Nt));
  cgxm=np.zeros((Ns,Nt));cgym=np.zeros((Ns,Nt))
  phim=np.zeros((Ns,Nt)); hm=np.zeros((Ns,Nt))

  kl=np.zeros((Ns+1,Nt)); # kl est défini ici

  edlxm=np.zeros((Ns,Nt));edlym=np.zeros((Ns,Nt))
  Em=np.zeros((Ns,Nt));

  kl=np.sqrt(kxl**2+kyl**2); # kl est calculé ici

  # calculs aux points intermediaires
  xm=.5*(xl[0:Ns,:]+xl[1:Ns+1,:]);
  ym=.5*(yl[0:Ns,:]+yl[1:Ns+1,:]);
  kxm=.5*(kxl[0:Ns,:]+kxl[1:Ns+1,:]);
  kym=.5*(kyl[0:Ns,:]+kyl[1:Ns+1,:]);
  km=.5*(kl[0:Ns,:]+kl[1:Ns+1,:]);
  cgxm=.5*(cgxl[0:Ns,:]+cgxl[1:Ns+1,:]);
  cgym=.5*(cgyl[0:Ns,:]+cgyl[1:Ns+1,:]);
  phim=.5*(phil[0:Ns,:]+phil[1:Ns+1,:]);
  hm=.5*(hl[0:Ns,:]+hl[1:Ns+1,:]);


  dxm=xl[1:Ns+1,:]-xl[0:Ns,:];
  dym=yl[1:Ns+1,:]-yl[0:Ns,:];
  edlxm=-dym; edlym=dxm;

  # Initialisation de la hauteur de houle
  Em[:,0]=np.zeros(Ns)+E0
  for i in range(1,Nt):
    cgdelim1=cgxm[:,i-1]*edlxm[:,i-1]+cgym[:,i-1]*edlym[:,i-1]
    cgdeli=cgxm[:,i]*edlxm[:,i]+cgym[:,i]*edlym[:,i]
    Em[:,i]=Em[:,i-1]*cgdelim1/cgdeli

  # visualisation hauteur de la houle Hm
  Em=np.where(Em>0, Em, 0)
  Hm=2*np.sqrt(Em);
  Hmc=.14*(2*pi)/km*np.tanh(km*hm) # critere de Miche
  Hm=np.where(Hm < Hmc, Hm, 0)
  contours = plt.contour(xm,ym,Hm,levels=30,colors='brown')  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Phase
  #########
  contours = plt.contour(xl,yl,phil/(2*pi),levels=20,colors='green',linewidths=2)  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Rayons
  #########
  Nsr=Ns//Nr
  for n in range(Ns+1)[::Nsr]:
    # trace des rayons
    xr=xl[n,:]; yr=yl[n,:];
    plt.plot(xr,yr,color="blue",linewidth=1)
    # trace des vitesses de groupe
    Ntq=Nt//10;
    xq=xl[n,::Ntq]; yq=yl[n,::Ntq]; kxq=kxl[n,::Ntq]; kyq=kyl[n,::Ntq];
    cgxq=cgxl[n,::Ntq]; cgyq=cgyl[n,::Ntq]
    plt.quiver(xq,yq,cgxq,cgyq,color='red',scale=100)
    plt.axis('equal')

  ##############
  plt.xlim(xmin,xmax)
  plt.ylim(ymin,ymax)
  fig.savefig('visurayons05.pdf')
  fig.savefig('visurayons05.jpg')

  return xl,yl,kxl,kyl,kl,cgxl,cgyl,phil,hl, \
         xm,ym,kxm,kym,km,cgxm,cgym,phim,hm,Hm

def visupara_05():
    # tracé de la longeur d'onde L, les vitesses de groupe et de phase, la hauteur H, l'élevation de la surface et la profondeur le long de rayons. B
    #   basé sur la fonction de la visualisation de la longueur d'onde visuLong_L02()

    # Extraction de Nrv rayons parmi Ns
    Nsv=Ns//Nrv;
    # intialisation des tableaux 2D
    scu=np.zeros((Nrv+1,Nt));
    Lr=np.zeros((Nrv+1,Nt));
    profondeur = np.zeros((Nrv+1, Nt))
    vitesses_phase = np.zeros((Nrv+1, Nt))
    vitesses_groupe = np.zeros((Nrv+1, Nt))
    hauteurs_houle = np.zeros((Nrv+1, Nt))
    elevations_surface = np.zeros((Nrv+1, Nt))
    pulsations = np.zeros((Nrv+1, Nt))

    # Extraction de Nsv rayons
    for n in range(Ns+Nsv)[::Nsv]:
        nr=n//Nsv  #print("n=",n,"nr=",nr)

        # coordonnée curviligne du rayon
        xrn=xl[n,:]; yrn=yl[n,:];
        for i in range(Nt-1):
            dx=xrn[i+1]-xrn[i]; dy=yrn[i+1]-yrn[i];
            scu[nr,i+1]=scu[nr,i]+np.sqrt(dx**2+dy**2);

        # cacul de k et L
        kxrn=kxl[n,:]; kyrn=kyl[n,:];
        krn=np.sqrt(kxrn**2+kyrn**2);
        Lr[nr,:]=2*pi/krn;


        # Profondeur
        profondeur[nr,:] = hl[n,:]
        # Vitesses et pulsation
        omega, c_phase, c_groupe = houle(krn, profondeur[nr,:])
        pulsations[nr,:] = omega
        vitesses_phase[nr,:] = c_phase
        vitesses_groupe[nr,:] = c_groupe
        # Hauteur et élévation
        idx_m = n if n < Ns else Ns - 1
        hauteur_a = Hm[idx_m, :]
        phase_a = phim[idx_m, :]
        hauteurs_houle[nr,:] = hauteur_a
        elevations_surface[nr,:] = (hauteur_a / 2) * np.cos(phase_a)


    # Initialisation de la figure pour chaque variable.
    fig, axe = plt.subplots(3, 2, figsize=(15, 12))
    for n in range(Nrv+1):
        # Longueur d'onde
        line,=axe[0,0].plot(scu[n,:],Lr[n,:],linewidth=1)
        Nnum=2*Nt//3; col=line.get_color() # indice où écrire le numéro
        axe[0, 0].text(scu[n,Nnum], Lr[n, Nnum],str(n), color=col, fontsize=12, va='center')

        # Vitesses de groupe et de phases + Le label soit la légende dans la boucle nous donne 11 légendes... donc méthode après
        axe[0, 1].plot(scu[n,:], vitesses_groupe[n,:], color=col, linewidth=1, linestyle='-')
        axe[0, 1].plot(scu[n,:], vitesses_phase[n,:], color=col, linewidth=1, linestyle=(0, (5, 1)))
        # Hauteur de la houle
        axe[1, 0].plot(scu[n,:], hauteurs_houle[n,:], color=col, linewidth=1)

        # Profondeur h
        axe[1, 1].plot(scu[n,:], profondeur[n,:], color=col, linewidth=1)

        # Variation de la pulsation le long des rayons
        omega_0 = pulsations[n, 0] # La valeur au tout premier point
        erreur_relative = (pulsations[n,:] - omega_0) / omega_0

        axe[2, 1].plot(scu[n,:], erreur_relative, color=col, linewidth=1)

    # Inversion de l'axe des ordonnées pour la hauteur de la houle.
    axe[1, 0].yaxis.set_inverted(True)

    # Elevation de la surface pour le rayon nº4
    axe[2, 0].plot(scu[4,:], elevations_surface[4,:], color='blue', linewidth=1)

    # Légendes des courbes
    axe[0, 0].set_title(r"Longueur d'onde $L$"); axe[0, 0].set_ylabel(r"$L$ (m)"); axe[0, 0].grid(True)
    axe[0, 1].set_title(r"Vitesses $c_\varphi$ et $c_g$"); axe[0, 1].set_ylabel("Vitesse (m/s)"); axe[0, 1].grid(True)
    # Pour légender les deux courbes des vitesses de phase et de groupe, on trace donc deux courbes invisible mais avec cette légende
    axe[0, 1].plot([], [], color='black', linestyle='-', label='Vitesse de groupe')
    axe[0, 1].plot([], [], color='black', linestyle=(0, (5, 1)), label='Vitesse de phase')
    axe[0, 1].legend(loc='lower left')
    axe[1, 1].set_title(r"Profondeur $h$"); axe[1, 0].set_ylabel(r"$h$ (m)"); axe[1, 0].invert_yaxis(); axe[1, 0].grid(True)
    axe[1, 0].set_title(r"Hauteur de la houle $H$"); axe[1, 1].set_ylabel(r"$H$ (m)"); axe[1, 1].grid(True)
    axe[2, 0].set_title(r"Élévation de la surface"); axe[2, 0].set_ylabel(r"$\eta$ (m)"); axe[2, 0].grid(True)
    axe[2, 1].set_title(r"Erreur relative (($\omega - \omega_0)$/$\omega_0$"); axe[2, 1].set_ylabel(r"Erreur $(\omega - \omega_0)/\omega_0$"); axe[2, 1].grid(True)


    fig.savefig('visupara_05.pdf')
```

```python
def visubathy3D_07(x,y):
  # Trace de  Zf=-h(x,y) en 3D
  # Sans modification de la fonction sauf le nom du fichier pour visualiser toutes les bathys des expériences suivantes et/ou précedantes
  X, Y = np.meshgrid(x, y)
  h,Hx,Hy=bathy(X,Y)
  Zf=-h
  # Color map
  # plot 3D
  fig = plt.figure(figsize=(9, 9))
  ax = fig.add_subplot(111, projection='3d')
  # clip Zf
  Zf=np.where(Zf<0, Zf,0)
  Nyh=0;
  Xtron=np.zeros((Nyh,Nx));
  Ytron=np.zeros((Nyh,Nx));
  Zftron=np.zeros((Nyh,Nx));
  Xtron=X[Nyh:Ny,:]; Ytron=Y[Nyh:Ny,:]; Zftron=Zf[Nyh:Ny,:]
  ax.plot_wireframe(Xtron, Ytron, Zftron, color='k', linewidth=0.5)
  ax.view_init(elev=20, azim=-40)  # tilt elev, rotate azim around z-axis
  ax.set_xlim(xmin, xmax)
  ax.set_ylim(ymin, ymax)
  #ax.set_zlim(-15, 0)
  #fig.colorbar(surf, ax=ax, shrink=.6, label=r'$Z_f(x, y)$')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel(r'$Z_f$')
  #ax.set_title(r'Batymétrie $Z_f(x, y)$')
  fig.savefig('visubathy3D_07.pdf')
  return fig

def bathy07(x,y):

    x_centre = 1000
    y_centre = -100
    bosse = 67.0
    h = hlarge - bosse * np.exp(-((x - x_centre)**2 + (y - y_centre)**2) / 300000)
    hx = (bosse/300000)* 2 * (x-x_centre) * np.exp(-((x - x_centre)**2 + (y - y_centre)**2) / 300000)
    hy = (bosse/300000)* 2 * (y-y_centre) * np.exp(-((x - x_centre)**2 + (y - y_centre)**2) / 300000)
    return h,hx,hy


def visurayons_miche():
  # Visualisation des rayons, adaptation et modification de la fonction visurayons()

  # plots de la bathymetrie et de la ligne d'emission L
  fig=visubathy(x,y);
  plt.plot([xd,xd+2*D],[yd,yd-2*D*np.sin(th)],color="green",linewidth=3)

  # Calcul de tous les rayons
  xl,yl,kxl,kyl,cgxl,cgyl,phil,hl=rayons(th)

  # Energie
  #########
  # initialisation des champs intermediaires
  xm=np.zeros((Ns,Nt));ym=np.zeros((Ns,Nt))
  kxm=np.zeros((Ns,Nt));kym=np.zeros((Ns,Nt))
  km=np.zeros((Ns,Nt));
  cgxm=np.zeros((Ns,Nt));cgym=np.zeros((Ns,Nt))
  phim=np.zeros((Ns,Nt)); hm=np.zeros((Ns,Nt))

  kl=np.zeros((Ns+1,Nt)); # kl est défini ici

  edlxm=np.zeros((Ns,Nt));edlym=np.zeros((Ns,Nt))
  Em=np.zeros((Ns,Nt));

  kl=np.sqrt(kxl**2+kyl**2); # kl est calculé ici

  # calculs aux points intermediaires
  xm=.5*(xl[0:Ns,:]+xl[1:Ns+1,:]);
  ym=.5*(yl[0:Ns,:]+yl[1:Ns+1,:]);
  kxm=.5*(kxl[0:Ns,:]+kxl[1:Ns+1,:]);
  kym=.5*(kyl[0:Ns,:]+kyl[1:Ns+1,:]);
  km=.5*(kl[0:Ns,:]+kl[1:Ns+1,:]);
  cgxm=.5*(cgxl[0:Ns,:]+cgxl[1:Ns+1,:]);
  cgym=.5*(cgyl[0:Ns,:]+cgyl[1:Ns+1,:]);
  phim=.5*(phil[0:Ns,:]+phil[1:Ns+1,:]);
  hm=.5*(hl[0:Ns,:]+hl[1:Ns+1,:]);

  dxm=xl[1:Ns+1,:]-xl[0:Ns,:];
  dym=yl[1:Ns+1,:]-yl[0:Ns,:];
  edlxm=-dym; edlym=dxm;

  # Initialisation de la hauteur de houle
  Em[:,0]=np.zeros(Ns)+E0
  for i in range(1,Nt):
    cgdelim1=cgxm[:,i-1]*edlxm[:,i-1]+cgym[:,i-1]*edlym[:,i-1]
    cgdeli=cgxm[:,i]*edlxm[:,i]+cgym[:,i]*edlym[:,i]
    Em[:,i]=Em[:,i-1]*cgdelim1/cgdeli

  # visualisation hauteur de la houle Hm
  Em=np.where(Em>0, Em, 0)
  Hm=2*np.sqrt(Em);
  Hmc=.14*(2*pi)/km*np.tanh(km*hm) # critere de Miche

  # AJOUT du calcul des coordonnées du déferlement
  x_def = []
  y_def = []
  for n in range(Ns):
      for i in range(Nt):
          if hm[n, i] > 0 and Hm[n, i] >= Hmc[n, i]:
              x_def.append(xm[n, i])
              y_def.append(ym[n, i])
              break # Point de rupture.

  Hm=np.where(Hm < Hmc, Hm, 0)
  contours = plt.contour(xm,ym,Hm,levels=30,colors='brown')  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Phase
  #########
  contours = plt.contour(xl,yl,phil/(2*pi),levels=20,colors='green',linewidths=2)  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Rayons
  #########
  Nsr=Ns//Nr
  for n in range(Ns+1)[::Nsr]:
    # trace des rayons
    xr=xl[n,:]; yr=yl[n,:];
    plt.plot(xr,yr,color="blue",linewidth=1)
    # trace des vitesses de groupe
    Ntq=Nt//10;
    xq=xl[n,::Ntq]; yq=yl[n,::Ntq]; kxq=kxl[n,::Ntq]; kyq=kyl[n,::Ntq];
    cgxq=cgxl[n,::Ntq]; cgyq=cgyl[n,::Ntq]
    plt.quiver(xq,yq,cgxq,cgyq,color='red',scale=100)

  # AJOUT: Tracé de la ligne de déferlement.
  if len(x_def) > 0:
      # On trace la ligne de déferlement par-dessus tout le reste (zorder=5)
      plt.plot(x_def, y_def, color='black', linewidth=3, zorder=5, label="Ligne de déferlement")
      plt.legend(loc='lower right')

  ##############
  plt.axis('equal')
  plt.xlim(xmin,xmax)
  plt.ylim(ymin,ymax)
  fig.savefig('visurayons_def.pdf')
  fig.savefig('visurayons_def.jpg')
  plt.show()

  return xl,yl,kxl,kyl,kl,cgxl,cgyl,phil,hl, \
         xm,ym,kxm,kym,km,cgxm,cgym,phim,hm,Hm

def visu_3D_rayons_elevation():
    # Création de la figure 3D
    fig = plt.figure(figsize=(6, 7))
    ax = fig.add_subplot(111, projection='3d')

    Nsv=Ns//Nrv;
    # intialisation des tableaux 2D
    scu=np.zeros((Nrv+1,Nt));
    profondeur = np.zeros((Nrv+1, Nt))
    vitesses_phase = np.zeros((Nrv+1, Nt))
    vitesses_groupe = np.zeros((Nrv+1, Nt))
    hauteurs_houle = np.zeros((Nrv+1, Nt))
    elevations_surface = np.zeros((Nrv+1, Nt))
    pulsations = np.zeros((Nrv+1, Nt))

    pas_points = max(1, Nt // 500)

    # Extraction de Nsv rayons
    for n in range(Ns+Nsv)[::Nsv]:
        nr=n//Nsv  #print("n=",n,"nr=",nr)

        # coordonnée curviligne du rayon
        xrn=xl[n,:]; yrn=yl[n,:];
        # cacul de k et L
        kxrn=kxl[n,:]; kyrn=kyl[n,:];
        krn=np.sqrt(kxrn**2+kyrn**2);

        # Profondeur (new)
        profondeur[nr,:] = hl[n,:]

        #Vitesses et pulsation
        omega, c_phase, c_groupe = houle(krn, profondeur[nr,:])
        pulsations[nr,:] = omega
        vitesses_phase[nr,:] = c_phase
        vitesses_groupe[nr,:] = c_groupe
        # Hauteur et élévation
        idx_m = n if n < Ns else Ns - 1
        hauteur_a = Hm[idx_m, :]
        phase_a = phim[idx_m, :]
        hauteurs_houle[nr,:] = hauteur_a
        elevations_surface[nr,:] = (hauteur_a / 2) * np.cos(phase_a)
        eta = (hauteur_a / 2) * np.cos(phase_a)

        ax.plot(xrn[::pas_points], yrn[::pas_points], eta[::pas_points], linewidth=1.5)
        # Tracé de la ligne 3D pour ce rayon spécifique

    X, Y = np.meshgrid(x, y)
    h,Hx,Hy=bathy07(X,Y)
    # plot 3D
    ax.set_xlabel('x (m)')
    ax.set_ylabel('y (m)')
    ax.set_zlabel(r'Élévation $\eta$ (m)')
    ax.set_zlim(-1.5, 1.5)
    ax.view_init(elev=35, azim=-55)

    plt.tight_layout()
    fig.savefig('exp07_3d.pdf')
    plt.show()
```

###      Méthode basée et inspirée par [Animated 3D random walk](https://matplotlib.org/stable/gallery/animation/random_walk.html) et le [Jupyter notebook de l'itération de Colebrook](https://colab.research.google.com/drive/1vSk-K2Bd_g0Cji8Y0dVQq2N_3ACr0hHf?usp=sharing)

Dans cette dernière experience, on crée une animation 3D pour illustrer la propagation et l'arrivée progressive des vagues vers le littoral à l'aide d'un masque.

### 07bis.a Nouveaux sous-programmes

```python
from matplotlib.animation import FuncAnimation, PillowWriter


def animation_vague():
    # initialisation d'une figure
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    Nrv = 5
    Nsv = Ns // Nrv
    pas = max(1, Nt//100)
    lignes_3d = []
    donnees_rayons = []

    # Frame 1
    for n in range(Ns+Nsv)[::Nsv]:

        nr=n//Nsv  #print("n=",n,"nr=",nr)

        # coordonnée curviligne du rayon
        xrn = xl[n, ::pas]; yrn = yl[n, ::pas]
        # cacul de k et L
        kxrn=kxl[n, ::pas]; kyrn=kyl[n, ::pas];
        krn=np.sqrt(kxrn**2+kyrn**2);
        hr = hl[n, ::pas]

        omega, _, _ = houle(krn, hr)

        idx_m = n if n < Ns else Ns - 1

        hauteur_a = Hm[idx_m, ::pas]
        phase_a = phim[idx_m, ::pas]
        eta = (hauteur_a / 2) * np.cos(phase_a)


        # On dessine la ligne et on la sauvegarde dans notre liste
        ligne, = ax.plot(xrn, yrn, eta, linewidth=1.5)
        lignes_3d.append(ligne)

        # On garde les données en mémoire pour le calcul temporel
        donnees_rayons.append((xrn, yrn, hauteur_a, phase_a, omega))
    # Mise en forme
    ax.set_zlim(-1.5, 1.5)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.view_init(elev=35, azim=-50)

    # 3. Paramètres de l'animation
    omega_moyen = np.nanmean(donnees_rayons[0][4]) # Calcul de la moyenne
    T_moyen = 2 * np.pi / omega_moyen # La période de la vague
    nb_frames_apparition = 30  # Arrivée de la houle
    nb_frames_total = 60
    temps_t = np.linspace(0, T_moyen, nb_frames_total) # Du temps 0 jusqu'à 1 période



    def update(frame_index):
        ''' Mise à jour de l'animation avec l'astuce de cacher la partie si la vague n'est pas encore arrivée'''
        t = temps_t[frame_index]
        #visible = (idx + 1) / nb_frames
        nb_points = 0

        if frame_index < nb_frames_apparition:
            # Vague en cours.
            partie_visible = (frame_index + 1) / nb_frames_apparition
        else:
            # Vague arrivée
            partie_visible = 1.0

        for i, ligne in enumerate(lignes_3d):
            xrn, yrn, hauteur_a, phase_a, omega_ray = donnees_rayons[i]

            # On recalcule eta avec le temps t
            eta_t = (hauteur_a / 2) * np.cos(phase_a - omega_ray * t)

            nb_points = len(eta_t)
            indice_limite = int(partie_visible * nb_points)

            eta_t_masque = np.copy(eta_t)
            eta_t_masque[indice_limite:] = np.nan


            ligne.set_data_3d(xrn, yrn, eta_t_masque)

        return lignes_3d

    anim = FuncAnimation(fig, update, frames=nb_frames_total)
    anim.save('Exp07_animation_vagues.gif', writer=PillowWriter(fps=15))
    plt.show()

```
